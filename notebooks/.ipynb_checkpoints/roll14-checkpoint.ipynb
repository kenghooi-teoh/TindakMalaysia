Get Age & Age Group

```python
def get_age(born):
    electionday = pd.to_datetime('2018-05-09')
    return electionday.year - born.year - ((electionday.month, electionday.day) < (born.month, born.day))
```

```python
roll14['Age'] = pd.to_datetime(roll14['TahunLahir']).apply(lambda x: get_age(x))
roll14['AgeGroup'] = pd.cut(roll14['Age'], [20,30,40,50,60,70,80,max(roll14['Age'])] )
```

```python
from math import*
  
def jaccard_similarity(x,y):
  
 intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
 union_cardinality = len(set.union(*[set(x), set(y)]))
 return intersection_cardinality/float(union_cardinality)
```

```python
def clean_words2(x):
    return x.upper()\
          .replace(',',' ')\
          .replace('NO',' ')\
          .replace('BLOCK',' ')\
          .replace('BLOK',' ')\
          .replace('JALAN','JLN')\
          .replace('KAMPUNG','KG')\
          .replace('TAMAN','TMN')\
          .replace('LORONG','LRG')\
          .replace('KUALA LUMPUR','KL')\
          .replace('SUNGAI','SG')\
          .replace('NAN','')\
          .replace(';',' ')\
          .replace('.',' ')\
          .replace('-','')\
          .replace('/',' ')\
          .replace('(',' ')\
          .replace(')',' ')\
          .replace('   ','  ')\
          .replace('  ',' ')\
          .strip()\
          .replace('  ',' ')
         
```

```python
addr_check['tokens_14'] = addr_check['Alamat_Rumah'].apply(lambda x: set(clean_words2(x).split()))
addr_check['tokens_13'] = addr_check['Alamat_SPR'].apply(lambda x: set(clean_words2(x).split()))
```

```python
addr_check['jaccard_similarity'] = addr_check[['tokens_14','tokens_13']].apply(lambda x: jaccard_similarity(x['tokens_14'],x['tokens_13']), axis=1)
```

```python
addr_check.drop(['tokens_14','tokens_13'], axis=1).to_csv('address_check2.csv')
```
