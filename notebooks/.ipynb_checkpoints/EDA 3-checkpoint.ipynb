```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from analysis import *
```

```python
# VOTERS
roll14 = pd.read_csv('Lembah Pantai GE14 Roll.csv')
roll13 = pd.read_csv('Lembah Pantai GE13 Roll.csv')

# SCORE
score_ppr = pd.read_excel('results.xlsx', sheetname='PPR')
score_ppr14 = score_ppr[score_ppr['GE'] == 'GE14']
score_ppr13 = score_ppr[score_ppr['GE'] == 'GE13']
```

```python
ppr_dms = ['BUKIT BANGSAR',
'BUKIT KERINCHI',
'KAMPUNG HAJI ABDULLAH HUKUM',
'KAMPUNG PANTAI HALT',
'TAMAN BUKIT ANGKASA',
'PANTAI HILL PARK']
```

```python
#______________________________________________________________________________________________________________________________

roll_ppr14 = roll14[(roll14['NamaDM']=='BUKIT BANGSAR') | (roll14['NamaDM']=='BUKIT KERINCHI') |(roll14['NamaDM']=='KAMPUNG HAJI ABDULLAH HUKUM') |(roll14['NamaDM']=='KAMPUNG PANTAI HALT')|(roll14['NamaDM']=='TAMAN BUKIT ANGKASA')|(roll14['NamaDM']=='PANTAI HILL PARK')]
roll_ppr13 = roll13[(roll13['NamaDM']=='BUKIT BANGSAR') | (roll13['NamaDM']=='BUKIT KERINCHI') |(roll13['NamaDM']=='KAMPONG HAJI ABDULLAH HUKOM') |(roll13['NamaDM']=='KAMPONG PANTAI HALT')|(roll13['NamaDM']=='TAMAN BUKIT ANGKASA')|(roll13['NamaDM']=='PANTAI HILL PARK')]

def calculate_age(born):
    electionday = pd.to_datetime('2018-05-09')
    return electionday.year - born.year - ((electionday.month, electionday.day) < (born.month, born.day))

roll_ppr14['Age'] = pd.to_datetime(roll_ppr14['TahunLahir']).apply(lambda x: calculate_age(x))
max_age = max(roll_ppr14['Age'].max(),roll_ppr13['Umur'].max())

roll_ppr14['AgeGroup'] = pd.cut(roll_ppr14['Age'], [20,30,40,50,60,70,80,max_age] )
roll_ppr13['AgeGroup'] = pd.cut(roll_ppr13['Umur'], bins=[20,30,40,50,60,70,80,max_age])
```

______

```python
ppr_age_compare = pd.DataFrame([roll_ppr13['AgeGroup'].value_counts(),roll_ppr14['AgeGroup'].value_counts() ]).T
ppr_age_compare.columns = ['GE13','GE14']

ppr_age_compare['GE13_p'] = ppr_age_compare['GE13']/ppr_age_compare['GE13'].sum()
ppr_age_compare['GE14_p'] = ppr_age_compare['GE14']/ppr_age_compare['GE14'].sum()

ppr_age_compare_p = ppr_age_compare[['GE13_p','GE14_p']]

ppr_age_compare2 = ppr_age_compare[['GE13','GE14']].unstack().reset_index()
ppr_age_compare2.columns = ['GE','AgeGroup','count']

ppr_age_compare_p2 = ppr_age_compare_p[['GE13_p','GE14_p']].unstack().reset_index()
ppr_age_compare_p2.columns = ['GE','AgeGroup','count']
```

```python
f,ax = plt.subplots(1,2, figsize=(12,5))
p1 = sns.barplot(y='count', x='AgeGroup', hue='GE',data=ppr_age_compare2, ax=ax[0])
p2 = sns.barplot(y='count', x='AgeGroup', hue='GE',data=ppr_age_compare_p2, ax=ax[1])

for patch in p1.patches:
    try:
        if patch.get_height()>0:
            p1.annotate(' %.f' % patch.get_height(), xy=( patch.get_x() + patch.get_width()/2 ,patch.get_height()+ 100), ha='center',fontsize=8)
    except:
        pass

for patch in p2.patches:
    try:
        if patch.get_height()>0:
            p2.annotate(' %.2f' % patch.get_height(), xy=( patch.get_x() + patch.get_width()/2 ,patch.get_height()+0.005), ha='center',fontsize=8)
    except:
        pass
    
p1.set_ylim(0,len(roll_ppr14))
p2.set_ylim(0,1)
    
plt.tight_layout()
plt.suptitle('Age Group (Count, Percentage)')

```
