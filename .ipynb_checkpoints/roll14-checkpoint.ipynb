```python
import pandas as pd
import folium
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
roll14 = pd.read_csv('Lembah Pantai GE14 Roll.csv')
roll13 = pd.read_csv('Lembah Pantai GE13 Roll.csv')
```

Removing Empty columns

```python
pd.isnull(roll14).sum() - len(roll14)
```

```python
roll14.dropna(how='all', axis=1, inplace=True)
```

Removing Non-contributing columns (same value for all rows)

```python
for col in roll14.columns:
    if len(roll14[col].unique()) == 1:
        roll14.drop(col, axis=1, inplace=True)
```

```python
roll14[[ 'NamaLokaliti', 'NamaDM']]
```

```python
roll14['Jantina'].value_counts()
```

Get Age & Age Group

```python
def calculate_age(born):
    electionday = pd.to_datetime('2018-05-09')
    return electionday.year - born.year - ((electionday.month, electionday.day) < (born.month, born.day))
```

```python
roll14['Age'] = pd.to_datetime(roll14['TahunLahir']).apply(lambda x: calculate_age(x))
roll14['AgeGroup'] = pd.cut(roll14['Age'], [20,30,40,50,60,70,80,max(roll14['Age'])] )
```

```python
roll13['AgeGroup'] = pd.cut(roll13['Umur'], [20,30,40,50,60,70,80,max(roll13['Umur'])])
```
